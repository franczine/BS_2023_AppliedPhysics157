```python
# load libraries

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2
```

## 6.1 Erosion and Dilation

```python
# generate test images

a = np.zeros((13,13))

for i in range(4, 9):
    for j in range(4,9):
        a[i][j] = 1
        #a[4][i] = 1

b = np.zeros((16,16))

for i in range(3,13):
    b[3][i] = 1
    b[12][i] = 1
    b[i][3] = 1
    b[i][12] = 1
    b[4][i] = 1
    b[11][i] = 1
    b[i][4] = 1
    b[i][11] = 1
    
c = np.zeros((13,13))

for i in range(4,9):
    c[i][6] = 1
    c[6][i] = 1
    
d = np.zeros((19,19))

for i in range(3, 8):
    for j in range(7,12):
        d[j][i] = 1

for i in range(8,11):
    d[9][i] = 1
        
for i in range(11, 16):
    for j in range(7,12):
        d[j][i] = 1
```

```python
# generate kernels

k1 = cv2.getStructuringElement(cv2.MORPH_RECT, ksize = (2,2))
k2 = cv2.getStructuringElement(cv2.MORPH_RECT, ksize = (2,1))
k3 = cv2.getStructuringElement(cv2.MORPH_RECT, ksize = (5,5))
```

```python
# function to perform erosion

def plotErosion(img, name):
    
    axlim = len(img)
    
    fig, ax = plt.subplots(2, 4, figsize = (11, 6))
    
    ax[0][0].imshow(img, cmap = 'gray')
    ax[0][0].set_title('original image')
    ax[0][1].imshow(cv2.erode(img, k1), cmap = 'gray')
    ax[0][1].set_title('erosion by kernel 1')
    ax[0][2].imshow(cv2.erode(img, k2), cmap = 'gray')
    ax[0][2].set_title('erosion by kernel 2')
    ax[0][3].imshow(cv2.erode(img, k3), cmap = 'gray')
    ax[0][3].set_title('erosion by kernel 3')
    ax[1][1].imshow(k1, cmap = 'gray')
    ax[1][1].set_xlim([- axlim / 2, axlim / 2])
    ax[1][1].set_ylim([-axlim / 2, axlim / 2])
    ax[1][1].set_title('kernel 1: 2 x 2 ones')
    ax[1][2].imshow(k2, cmap = 'gray')
    ax[1][2].set_xlim([- axlim /2 , axlim/2])
    ax[1][2].set_ylim([-axlim /2 , axlim/2])
    ax[1][2].set_title('kernel 2: 2 x 1 ones')
    ax[1][3].imshow(k3, cmap = 'gray')
    ax[1][3].set_xlim([- axlim * 2 / 5, axlim * 3 / 5])
    ax[1][3].set_ylim([-axlim * 2 / 5, axlim * 3 /5 ])
    ax[1][3].set_title('kernel 3: 5 x 5 ones')
    
    for i in range(2):
        for j in range(4):
            ax[i][j].axis('off')
            #ax[i][j].set_xticks([])
            #ax[i][j].set_yticks([])
    
    ax[1][0].axis('off')
    
    plt.suptitle('erosion of a {}'.format(name), size = 15)
```

```python
# perform erosion on image a
plotErosion(a, '5 x 5 square')
plt.savefig('pics/erosion_a')
```

```python
# perform erosion on image b
plotErosion(b, '10 x 10 hollow square, 2 boxes thick')
plt.savefig('pics/erosion_b')
```

```python
# perform erosion on image c
plotErosion(c, 'plus sign, 1 box thick, 5 boxes along each line')
plt.savefig('pics/erosion_c')
```

```python
# perform erosion on image d
plotErosion(d, 'dumbbell - two 5x5 squares connected by a 3x1 line')
plt.savefig('pics/erosion_d')
```

```python
# function for performing dilation

def plotDilation(img, name):
    
    axlim = len(img)
    
    fig, ax = plt.subplots(2, 4, figsize = (12, 6))
    
    ax[0][0].imshow(img, cmap = 'gray')
    ax[0][0].set_title('original image')
    ax[0][1].imshow(cv2.dilate(img, k1), cmap = 'gray')
    ax[0][1].set_title('dilation by kernel 1')
    ax[0][2].imshow(cv2.dilate(img, k2), cmap = 'gray')
    ax[0][2].set_title('dilation by kernel 2')
    ax[0][3].imshow(cv2.dilate(img, k3), cmap = 'gray')
    ax[0][3].set_title('dilation by kernel 3')
    ax[1][1].imshow(k1, cmap = 'gray')
    ax[1][1].set_xlim([- axlim / 2, axlim / 2])
    ax[1][1].set_ylim([-axlim / 2, axlim / 2])
    ax[1][1].set_title('kernel 1: 2 x 2 ones')
    ax[1][2].imshow(k2, cmap = 'gray')
    ax[1][2].set_xlim([- axlim /2 , axlim/2])
    ax[1][2].set_ylim([-axlim /2 , axlim/2])
    ax[1][2].set_title('kernel 2: 2 x 1 ones')
    ax[1][3].imshow(k3, cmap = 'gray')
    ax[1][3].set_xlim([- axlim * 2 / 5, axlim * 3 / 5])
    ax[1][3].set_ylim([-axlim * 2 / 5, axlim * 3 /5 ])
    ax[1][3].set_title('kernel 3: 5 x 5 ones')
    
    for i in range(2):
        for j in range(4):
            ax[i][j].axis('off')
            #ax[i][j].set_xticks([])
            #ax[i][j].set_yticks([])
    
    ax[1][0].axis('off')
    
    plt.suptitle('dilation of a {}'.format(name), size = 15)
```

```python
# perform dilation on image a
plotDilation(a, '5 x 5 square')
plt.savefig('pics/dilation_a')
```

```python
# perform dilation on image b

plotDilation(b, '10 x 10 hollow square, 2 boxes thick')
plt.savefig('pics/dilation_b')
```

```python
# perform dilation on image c
plotDilation(c, 'plus sign, 1 box thick, 5 boxes along each line')
plt.savefig('pics/dilation_c')
```

```python
# perform dilation on image d
plotDilation(d, 'dumbbell - two 5x5 squares connected by a 3x1 line')
plt.savefig('pics/dilation_d')
```

## 6.2 Cleaning of Image Before Feature Extraction

```python
I = mpimg.imread('malaria(1).jpeg')
plt.axis('off')
plt.imshow(I)
```

```python
def rgb2gray(rgb):
    """
    converts rgb image to grayscale
    """
    return np.dot(rgb[...,:3], [0.2989, 0.5870, 0.1140])
```

```python
# convert image to gray
Igray = rgb2gray(I)
plt.axis('off')
plt.imshow(Igray, cmap = 'gray')
```

```python
hist, bins = np.histogram(Igray, bins=256, density=True)
```

```python
plt.plot(hist)
```

```python
# segment image
cells = np.copy(Igray)
cells[np.where((cells < 164) & (cells > 50))] = 1
```

```python
# segment image again to create binary values

segment = np.copy(cells)
segment[np.where(cells > 10)] = 0
```

```python
# cleaning image of cells

kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (3,3))
#step1 = cv2.erode(segment, kernel, iterations = 1)
step1 = cv2.morphologyEx(segment, cv2.MORPH_CLOSE, kernel, iterations = 1)
```

```python
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (3,3))
step2 = cv2.erode(step1, kernel, iterations = 3)
```

```python
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (2,2))
step3 = cv2.dilate(step2, kernel, iterations = 1)
```

```python
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (4,4))
step4 = cv2.erode(step3, kernel, iterations = 1)
```

```python
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (3,3))
step5 = cv2.dilate(step4, kernel, iterations = 3)
```

```python
fig, ax = plt.subplots(2, 3, figsize = (12, 6.5))

ax[0][0].imshow(segment, cmap = 'gray')
ax[0][0].set_title('original image')

ax[0][1].imshow(step1, cmap = 'gray')
ax[0][1].set_title('step 1: erosion')

ax[0][2].imshow(step2, cmap = 'gray')
ax[0][2].set_title('step 2: closing')

ax[1][0].imshow(step3, cmap = 'gray')
ax[1][0].set_title('step 3: erosion')

ax[1][1].imshow(step4, cmap = 'gray')
ax[1][1].set_title('step 4: closing')

ax[1][2].imshow(step5, cmap = 'gray')
ax[1][2].set_title('step 5: dilation')

plt.suptitle('Clean Up of Cell Images', size = 16)
plt.tight_layout()

for i in range(2):
    for j in range(3):
        ax[i][j].axis('off')
        
plt.savefig('pics/cells')
```

```python
fig, ax = plt.subplots(1,2, figsize = (12, 7))

ax[0].imshow(segment, cmap = 'gray')
ax[0].set_title('original image')
ax[0].axis('off')
ax[1].imshow(step5, cmap = 'gray')
ax[1].set_title('cleaned image')
ax[1].axis('off')

plt.savefig('pics/results')
```
