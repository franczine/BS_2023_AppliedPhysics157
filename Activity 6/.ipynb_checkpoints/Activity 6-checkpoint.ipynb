```python
# load libraries

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

## 6.2 

```python
I = mpimg.imread('malaria(1).jpeg')
plt.axis('off')
plt.imshow(I)
```

```python
def rgb2gray(rgb):
    """
    converts rgb image to grayscale
    """
    return np.dot(rgb[...,:3], [0.2989, 0.5870, 0.1140])
```

```python
Igray = rgb2gray(I)
plt.axis('off')
plt.imshow(Igray, cmap = 'gray')
```

```python
hist, bins = np.histogram(Igray, bins=256, density=True)
```

```python
plt.plot(hist)
```

```python
cells = np.copy(Igray)
cells[np.where((cells < 164) & (cells > 100))] = 0
#cells[np.where((cells > 164))] = 0
plt.imshow(cells)#, cmap = 'gray')
```

```python
import cv2
```

```python
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE , (1,1))
erosion = cv2.erode(cells, kernel, iterations = 2)
plt.imshow(erosion)
```

```python
cells = np.copy(Igray)
cells[np.where((cells < 164))] = 0
plt.imshow(cells)#, cmap)

kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (1,1))
dilated = cv2.dilate(cells, kernel, iterations = 4)
plt.imshow(dilated)
```
